## Parametros de la funcion _read_csv_

Documentacion: https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

Ahora, importemos archivo separado por punto y coma utilizando el parametro _sep_

```python
df_punto_coma = pd.read_csv('./superstore_data_punto_coma.csv', sep=';')
df_punto_coma.head()
```

Carguemos nuevamente el archivo separado por coma y leamos solo 5 registros con _nrows_

```python
df_5_rows = pd.read_csv('./superstore_data.csv', nrows=5)
df_5_rows
```

Carguemos nuevamente el archivo separado por coma y leamos solo algunas columnas con _usecols_

```python
df_seleccion = pd.read_csv('./superstore_data.csv', usecols=['Id', 'Year_Birth', 'Income'])
df_seleccion.head()
```

```python
# seleccion de columnas por su posicion
df_seleccion = pd.read_csv('./superstore_data.csv', usecols=[0, 1, 4])
df_seleccion.head()
```

## Importar archivos Excel

```python
import pandas as pd

archivo = 'emisiones_CO2.xlsx'
df = pd.read_excel(archivo)
df
```

Obtener el nombre de las hojas del archivo excel.

```python
pd.ExcelFile(archivo).sheet_names
```

## Parametros de la funcion _read_excel_

Documentacion: https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html

Vamos a abrir una hoja epecifica del archivo de excel con el parametro _sheet_name_. Por defecto, muestra la hoja con indice 0.

```python
# cargar hoja percapita
percapita = pd.read_excel(archivo, sheet_name='emisiones_percapita')
percapita.tail()
```

```python
# cargar hoja fuentes
fuentes = pd.read_excel(archivo, sheet_name='fuentes')
fuentes.sample(5)
```

Vamos a seleccionar una serie de columnas mediante intervalos, tal como se hace en excel a traves de letras como indices.

```python
intervalo = pd.read_excel(archivo, sheet_name='emisiones_C02', usecols='A:D')
intervalo.head()
```

Vamos a seleccionar una serie de registros, como la vez pasada, a traves de intervalos

```python
intervalo_2 = pd.read_excel(archivo, sheet_name='emisiones_C02', usecols='A:D', nrows=10)
intervalo_2
```

## Exportar archivo Excel

Vamos a exportar el df '_intervalo_' con el metodo _to_excel_.

```python
intervalo.to_excel('CO2_percapita.xlsx', index=False)
```

```python
# cargar archivo exportado
pd.read_excel('CO2_percapita.xlsx')
```

## Leyendo datos de Google Sheets

Obtenemos el link del archivo desde Google Sheets.

link: https://docs.google.com/spreadsheets/d/1P9-66nt13OmJjB-AjySSDw0YbvAZxEKZ/edit?usp=sharing&ouid=112995986981378980539&rtpof=true&sd=true

Para leer el archivo, debemos ubicar el ID que inicia despues del slash de 'd' y termina en el siguiente slash.

```python
sheet_id = '1P9-66nt13OmJjB-AjySSDw0YbvAZxEKZ'

# guardamos la url pero cambiando la extension sharing por gviz/tq?tqx=out:csv$sheet
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
```

```python
datos_google = pd.read_csv(url)
datos_google
```

Vamos a abrir una hoja especifica del archivo en Google Sheets

```python
sheet_id = '1P9-66nt13OmJjB-AjySSDw0YbvAZxEKZ'
sheet_name = 'emisiones_percapita'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'

percapita = pd.read_csv(url)
percapita.sample(5)
```
