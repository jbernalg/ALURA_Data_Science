## Importar archivos JSON

El archivo "_pacientes.json_" no posee diccionarios anidados en los valores por lo que no presenta problemas al cargar.

```python
import pandas as pd
```

```python
# leer archivo json
df_pacientes = pd.read_json('./pacientes.json')
df_pacientes
```

El archivo "_pacientes_2.json_" si presenta diccionarios anidados lo que hace que la visuializacion no sea la correcta.

```python
# leer archivo json con diccionarios anidados
df_pacientes_2 = pd.read_json('./pacientes_2.json')
df_pacientes_2
```

Para solucionarlo, debemos conocer la estructura del json y por tanto utilizamos el siguiente editor de json: 

- [JSONcrack](https://jsoncrack.com/editor) 

Al visualizar el archivo en este editor obtenemos un diagrama donde podemos identificar como se estructura el json



Ahora, procedemos a normalizar el archivo json mediante el metodo _json_normalize()_ que busca expandir cada diccionario y colocar las variables en nuevas columnas. Debemos pasarle la variable que contiene el diccionario anidado

```python
df_normalizado = pd.json_normalize(df_pacientes_2['Pacientes'])
df_normalizado
```

## Exportar archivo JSON

```python
df_normalizado.to_json('historico_pacientes_norm.json')
```

```python
# abrimos el archivo json
pd.read_json('historico_pacientes_norm.json')
```
