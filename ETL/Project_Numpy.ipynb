## Regresin Lineal

Las variables que conforman la ecuacin de la recta son:

$$\alpha = pendiente$$
$$\beta = interseccion$$
$$n = observaciones$$
$$y = precio$$
$$x = fecha$$


La solucin para la pendiente es:

$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

Y para la interseccin:
 $$\beta = \frac{\sum y - \alpha \sum x}{n} $$

Y obtenemos la ecuacin de la recta:

$$y = \alpha x + \beta $$

Vamos a calcular la pendiente de la recta que mejor se ajusta a los datos de los precios de Moscu, de forma analitica

```python
# observaciones
n = np.size(Moscu)

# precio
y = Moscu

# fechas
X = fechas
```

```python
# calcular la pendiente (alpha)
a = (n*np.sum(X*y) - np.sum(X)*np.sum(y))/(n*np.sum(X**2) - np.sum(X)**2)

# calcular interseccion (beta)
b = np.mean(y) - a*np.mean(X)
```

```python
print(f'La pendiente es: {a}\nLa interseccion es: {b}')
```

```python
# calculamos la recta
y = a*x + b
y
```

Vamos a hallar el valor absoluto de la diferencias entre los datos y la recta (conocido como el Error) pero esta vez con la mejor recta calculada a partir de los datos de Moscu

```python
np.linalg.norm(Moscu - y)
```

> De esta forma obtenemos el mnimo error posible al usar la recta que mejor se ajusta a los datos.

Si recapitulamos, vamos a comparar las diferentes rectas usadas junto a su respectivo error:

- (Recta Estimada 1): Pendiente = 1.5, Intereseccion = 78 ----> Error = 500.33

- (Recta Estimada 2): Pendiente = 0.5, Intereseccion = 78 ----> Error = 103.53

- (Recta mejor ajustada): Pendiente = 0.43, Intereseccion = 79.64 ----> Error = 101.79

Ahora, grafiquemos la recta que mejor se ajusta a los datos

```python
# Grafica de la mejor recta
plt.plot(fechas,Moscu)
plt.title('Precio vs fechas (Moscu)')
plt.plot(X, y)
plt.show()
```

## Valores Aleatorios

Numpy tiene la funcionalidad de generar numeros aleatorios los cuales son utiles en diversas reas. 

```python
# valores aleatorios enteros
np.random.randint(40, 100, 100)
```

```python
# valores aleatorioas decimal
np.random.uniform(0.1, 0.9, 100)
```

Supongamos que no tenemos a la mano la frmula para calcular la recta que mejor se ajusta a los datos y por tanto, debemos calcularla de forma manual probando con diferentes valores aleatorios decimales para la pendiente.

```python
# 100 valores aleatorios decimales entre 0.1 y 0.9 para la pendiente
pendientes = np.random.uniform(0.1, 0.9, 100)
```

```python
# creamos un array vacio para guardar los valores de norma generados
norma = np.array([])

for i in range(100):
    np.append(norma, np.linalg.norm(Moscu - (pendientes[i]*X + b)))

```

Los valores aleatorios generados por Numpy no vienen ordenados y tampoco se generan los mismos valores al ejecutar la aleatoriedad nuevamente.


A veces necesitamos reproducir los mismos nmeros aleatorios sobre todo cuando estamos modelando. Para ello, acudimos al concepto de 'Semillas' la cual define el estado de aleatoriedad de los nmeros permitiendo reproducir los mismos valores. Para que reproduzca los resultados de forma efectiva, la semilla debe establecerse en el mismo cdigo donde se generan los nmeros aleatorios.

```python
# definir semilla
np.random.seed(42)
# numeros aleatorios de un rango dado
np.random.uniform(0.1, 0.9, 5)
```

Sabiendo esto, vamos a calcular los valores de norma con los valores de pendientes creados aleatoriamente, esta vez usando una semilla,

```python
np.random.seed(42)

norma = np.array([])

for i in range(100):
    norma = np.append(norma, np.linalg.norm(Moscu - (pendientes[i]*X + b)))

norma
```

```python
# comprobemos la reproducibilidad de los resultados
np.random.seed(42)

norma = np.array([])

for i in range(100):
    norma = np.append(norma, np.linalg.norm(Moscu - (pendientes[i]*X + b)))

norma
```

> Vemos que los resultados en ambas iteraciones son los mismos para la norma. Esto comprueba la habilidad de la semillas para reproducir resultados.
