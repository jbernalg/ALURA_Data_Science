## Visualizacion y seleccion

```python
# obtener fechas
fechas = datos_transpuetos[:,0]
fechas
```

```python
# obtener precio
precios = datos_transpuetos[:,1:]
precios
```

```python
# libreria para visualizaciones
import matplotlib.pyplot as plt
```

```python
# graficar fecha vs precio
plt.plot(fechas, precios)
plt.show()
```

```python
# seleccionamos los precios de las manzanas de moscu
plt.plot(fechas, precios[:,0])
plt.show()
```
